```python

ds['Num_of_Loan'] = pd.to_numeric(ds['Num_of_Loan'], errors='coerce')  # Converts invalid entries to NaN
ds.dropna(subset=['Num_of_Loan'], inplace=True)                       # Drops rows with NaN values
ds['Num_of_Loan'] = ds['Num_of_Loan'].astype(int)                     # Converts valid values to integers
```

```python
ds['Annual_Income'] = pd.to_numeric(ds['Annual_Income'], errors='coerce')  
ds.dropna(subset=['Annual_Income'], inplace=True)                      
ds['Annual_Income'] = ds['Annual_Income'].astype(int)  
```

```python
ds['Changed_Credit_Limit'] = pd.to_numeric(ds['Changed_Credit_Limit'], errors='coerce')  
ds.dropna(subset=['Changed_Credit_Limit'], inplace=True)                       
ds['Changed_Credit_Limit'] = ds['Changed_Credit_Limit'].astype(float)  
```

```python
ds['Outstanding_Debt'] = pd.to_numeric(ds['Outstanding_Debt'], errors='coerce')  
ds.dropna(subset=['Outstanding_Debt'], inplace=True)                       
ds['Outstanding_Debt'] = ds['Outstanding_Debt'].astype(float)  
```

```python
ds['Amount_invested_monthly'] = pd.to_numeric(ds['Amount_invested_monthly'], errors='coerce')  
ds.dropna(subset=['Amount_invested_monthly'], inplace=True)                       
ds['Amount_invested_monthly'] = ds['Amount_invested_monthly'].astype(float) 
```

```python
ds['Monthly_Balance'] = pd.to_numeric(ds['Monthly_Balance'], errors='coerce')  
ds.dropna(subset=['Monthly_Balance'], inplace=True)                       
ds['Monthly_Balance'] = ds['Monthly_Balance'].astype(float)
```

```python
ds['Num_of_Delayed_Payment'] = pd.to_numeric(ds['Num_of_Delayed_Payment'], errors='coerce')
```

```python
ds.isna().sum()
```

```python
ds['Name'].fillna(ds['Name'].mode()[0], inplace=True)
ds['Type_of_Loan'].fillna(ds['Type_of_Loan'].mode()[0], inplace=True)
ds['Credit_Mix'].fillna(ds['Credit_Mix'].mode()[0], inplace=True)
```

```python
ds['Monthly_Inhand_Salary'].fillna(ds['Monthly_Inhand_Salary'].mean(), inplace=True)
ds['Num_of_Delayed_Payment'].fillna(ds['Num_of_Delayed_Payment'].median(), inplace=True)
ds['Changed_Credit_Limit'].fillna(ds['Changed_Credit_Limit'].mean(), inplace=True)
```

```python
import re
def convert_to_months(age):
     if isinstance(age, str):
         years, months = map(int, re.findall(r'\d+', age))
         return years * 12 + months
     return age
ds['Credit_History_Age'] = ds['Credit_History_Age'].apply(convert_to_months)
ds['Credit_History_Age'].fillna(ds['Credit_History_Age'].mean(), inplace=True)
```

```python
ds['Monthly_Balance'].fillna(
     ds['Annual_Income'] - ds['Total_EMI_per_month'], inplace=True)
```

```python
ds['Num_Credit_Inquiries'].fillna(
    ds.groupby('Credit_Mix')['Num_Credit_Inquiries'].transform('mean'), inplace=True)
```

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
ds['Payment_Behaviour']=label_encoder.fit_transform(ds['Payment_Behaviour'])
ds['Monthly_Balance']=label_encoder.fit_transform(ds['Monthly_Balance'])
ds['Credit_Score']=label_encoder.fit_transform(ds['Credit_Score'])
ds['Occupation']=label_encoder.fit_transform(ds['Occupation'])
ds['SSN']=label_encoder.fit_transform(ds['SSN'])
ds['Customer_ID']=label_encoder.fit_transform(ds['Customer_ID'])
ds['ID']=label_encoder.fit_transform(ds['ID'])
ds['Payment_of_Min_Amount']=label_encoder.fit_transform(ds['Payment_of_Min_Amount'])
ds['Credit_Mix']=label_encoder.fit_transform(ds['Credit_Mix'])
ds['Type_of_Loan']=label_encoder.fit_transform(ds['Type_of_Loan'])
```

```python
ds.drop(columns = ['Name','Month',],axis = 1, inplace = True)
```

```python
ds.drop(columns=['Age'],axis=1,inplace=True)
```

```python
X = ds.drop(columns = 'Credit_Score')
y = ds['Credit_Score']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=True, test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression  
logreg= LogisticRegression()  
logreg.fit(X_train, y_train)
```

```python
logreg.score(X_train,y_train)
```

```python
logreg.score(X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_class=DecisionTreeClassifier(criterion='entropy',random_state=0)
dt_class.fit(X_train,y_train)
```

```python
dt_class.score(X_train,y_train)
```

```python
dt_class.score(X_test,y_test)
```

```python
dt_class = DecisionTreeClassifier(
    criterion='entropy',     # Using 'entropy' for information gain
    random_state=0,          # Ensures reproducibility
    max_depth=5,             # Limits the tree depth
    min_samples_split=10,    # Minimum samples required to split a node
    min_samples_leaf=5,      # Minimum samples required in a leaf node
    max_leaf_nodes=15,       # Limits the number of leaf nodes
    ccp_alpha=0.01           # Post-pruning regularization parameter
)
# Fit the classifier to your training data
dt_class.fit(X_train, y_train)
# Predict on test data
y_pred = dt_class.predict(X_test)
```
